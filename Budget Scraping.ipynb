```python
def get_table(link):
    td = []
    movie_budget = requests.get(link)
    soup1 = BeautifulSoup(movie_budget.content, 'html.parser')
    for tr in soup1.find_all('tr')[1:]:
        movie_info = []
        tds = tr.find_all('td')
        for index,i in enumerate(tds):
            movie_info.append(tds[index].get_text())
        td.append(movie_info)
    return td
```

```python
get_table("https://www.the-numbers.com/movie/budgets/all/101")
```

```python
def make_links():
    links = []
    links.append('https://www.the-numbers.com/movie/budgets/all')
    for i in range(1,58):
        links.append(f'https://www.the-numbers.com/movie/budgets/all/{i}01')
    return links
```

```python
def get_prod_budgets(links):
    all_budgets = []
    for link in links:
        all_budgets.extend(get_table(link))
    return all_budgets
```

```python
prod_budgets = get_prod_budgets(make_links())
```

```python
get_table("https://www.the-numbers.com/movie/budgets/all")
```

```python
budget_df = pd.DataFrame(prod_budgets,columns=['num','release_date','title','production_budget','dom_gross','ww_gross'])
```
